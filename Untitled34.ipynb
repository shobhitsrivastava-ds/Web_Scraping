```python
#Importing the required library
from bs4 import BeautifulSoup
import requests
import csv
import pprint
import csv
```

```python
#Creating csv file to store data parsed
csv_file= open("imdb_rating.csv","w")
```

```python
csv_writer = csv.writer(csv_file)
```

```python
csv_writer.writerow(["S.no","Movie","Year","Rating"])
```

```python
#Making the http get request to the website
source = requests.get("https://www.imdb.com/chart/top").text
```

```python
source
```

```python
#Parsing the result
soup = BeautifulSoup(source,"lxml")
```

```python
#Seperating the required tab
details=soup.find("div",class_="lister")
```

```python
title=details.find_all("td",class_= "titleColumn")
rating = details.find_all("td",class_="ratingColumn imdbRating")
```

```python
for i in range(len(title)):
    tit=title[i].text.split("\n")[2].encode("utf-8").strip()
    year= title[i].text.split("\n")[3].encode("utf-8")
    rat=rating[i].text.split("\n")[1].encode("utf-8")
    csv_writer.writerow([i+1,tit,year,rat])
```
